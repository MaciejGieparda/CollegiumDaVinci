#### Duck typing

````python
def isiterable(obj):
    try:
        iter(obj)
        return True
    except TypeError: # not iterable
        return False
````

````python
isiterable('a string')
isiterable([1, 2, 3])
isiterable(5)
````

````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

p1 = Person("John", 36)

print(p1.name)
print(p1.age)
````

````python
class Male(Person):
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.gender = 'm'
````

````python
male = Male('Antonio', 45)
````

````python
isinstance(male, Male)
````

````python
isinstance(male, Person)
````
