```python
import re

line = "Pinguins are smarter than polar bears"

matchObj = re.match( r'(.*) are (.*?) .*', line, re.M|re.I)

if matchObj:
   print("matchObj.group() : " + matchObj.group())
   print("matchObj.group(1) : " + matchObj.group(1))
   print("matchObj.group(2) : " + matchObj.group(2))
else:
   print("No match!!")
```

```python
p = re.compile(r"(\d+),\s*(\d+),\s*(\d+)")
w = p.search('1,22, 333, a,2,b,c; ni,5,4,3,2,1')
w.group(0), w.group(1), w.group(2), w.group(3)
```

```python
p = re.compile(r"(\d+),\s*(\d+),\s*(\d+)")
w = p.finditer('1,22, 333, a,2,b,c; ni,5,4,3,2,1')
for i in w:
    print(i.group(0), i.group(1), i.group(2), i.group(3))
```

```python
p = re.compile(r"(?P<first>\d+),\s*(?P<second>\d+),\s*(?P<third>\d+)")
w = p.search('1,22, 333, a,2,b,c; ni,5,4,3,2,1')
w.group('first'), w.group('second'), w.group('third')
```

```python
numbers = "123-123 456-789 210-012"
for i in re.finditer(r"(?P<number>\d+)-(?P=number)", numbers):
    print(i.group(0))
```
