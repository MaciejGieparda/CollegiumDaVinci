```python
def popraw_nazwy_kolumn(df):
    # Zamiana nazw kolumn na małe litery
    df.rename(columns=lambda x: x.lower(), inplace=True)
    
    # Zamiana polskich znaków na łacińskie litery
    try:
        df.columns = [col.encode('latin1').decode('unicode_escape').encode('latin1').decode('ascii', 'ignore') for col in df.columns] 
    except:   
        df.columns = [col.translate(str.maketrans("ąćęłńóśźż", "acelnoszz")) for col in df.columns]
    
    try:
        df.drop(columns=['brana'], inplace=True)
    except:
        pass

    # Poprawianie specyficznych nazw kolumn
    df.rename(columns={
        'ceni': 'plec',
        'wykszta': 'wyksztalcenie',
        'roczny_dochd': 'roczny_dochod',
        'wydatki_ywno': 'wydatki_zywnosc',
        'osoby_zalene': 'osoby_zalezne',
        'oszczdnoci': 'oszczednosci',
        'inwestycje_nieruchomoci': 'inwestycje_nieruchomosci',
        'lata_dowiadczenia': 'lata_doswiadczenia'
    }, inplace=True)



    return df

def czyszczenie_danych(df):
    df = df.applymap(lambda x: x.translate(str.maketrans("ąćęłńóśźż", "acelnoszz")) if isinstance(x, str) else x)
    df['wyksztalcenie'] = df['wyksztalcenie'].replace({'WyÅ¼sze': 'Wyzsze','Wy¿sze': 'Wyzsze', 'Å\x9arednie': 'Srednie', 'Tehnik': 'Technik', 'Średnie': 'Srednie'})
    df['stanowisko'] = df['stanowisko'].replace({'InÅ¼ynier': 'Inzynier'})
    df = df[df['wiek'] < 100]
    #zamien na integer
    df['karty_kredytowe'] = df['karty_kredytowe'].replace({'-': 0})
    df['karty_kredytowe'] = pd.to_numeric(df['karty_kredytowe'], errors='coerce').fillna(0).astype(int)
    
    return df

df = popraw_nazwy_kolumn(df)
df2 = popraw_nazwy_kolumn(df2)

df = pd.concat([df, df2], ignore_index=True)
df = czyszczenie_danych(df)
df.head()
```

Analiza demograficzna

```python
# Analiza wieku
average_age = df['wiek'].mean()
plt.figure(figsize=(12, 6))
plt.hist(df['wiek'], bins=30)
plt.title('Rozkład wieku')
plt.show()

# Analiza płci
gender_distribution = df['plec'].value_counts()
plt.figure(figsize=(12, 6))
plt.hist(df['plec'])
plt.title('Rozkład płci')
plt.show()

# Analiza wykształcenia
education_distribution = df['wyksztalcenie'].value_counts()
plt.figure(figsize=(12, 6))
plt.hist(df['wyksztalcenie'])
plt.title('Rozkład poziomu wykształcenia')
plt.show()

# Wyświetlanie wyników
print("Średni wiek:", average_age)
print("Najczestrze wyksztalcenie:", education_distribution.idxmax())
```

##### Analiza Zawodowa

```python
job_distribution = df['stanowisko'].value_counts()
print("Najczestrzy zawod:", job_distribution.idxmax())
plt.figure(figsize=(12, 6))
job_distribution.plot(kind='bar', figsize=(10, 6))
plt.title('Rozkład stanowisk')
plt.xlabel('Stanowisko')
plt.ylabel('Liczba pracowników')
plt.show()
```

##### Stosunek wydatków do dochodu

```python
expenses = df['wydatki_mieszkanie'] + df['wydatki_zywnosc'] + df['wydatki_transport'] + df['wydatki_edukacja']
income = df['roczny_dochod']
df['ratio_expenses_income'] = expenses / income

#Wykresy
plt.figure(figsize=(12, 6))
sns.boxplot(x='stanowisko', y='ratio_expenses_income', data=df)
plt.title('Rozkład wskaźnika wydatków do dochodów')
plt.show()
```

##### Analiza inwestycji

```python
investment = df['inwestycje_akcje'] + df['inwestycje_nieruchomosci']
df['investment_percentage'] = (investment / income)

#Wykreślenie wykresu
plt.figure(figsize=(12, 6))
sns.scatterplot(x='ratio_expenses_income', y='investment_percentage', data=df)
plt.title('Zależność między wydatkami a inwestycjami')
plt.xlabel('Stosunek wydatków do dochodu')
plt.ylabel('Procent inwestycji')
plt.show()
```

##### Wskaźnik oszczędności

```python
savings = df['oszczednosci']
df['savings_indicator'] = savings / income

#wykresy savings indicator
plt.figure(figsize=(12, 6))
df['savings_indicator'].plot(kind='hist', bins=30)
plt.title('Rozkład wskaźnika oszczędności')
plt.show()
```

##### Stosunek zadłużenia do dochodu

```python
debt = df['kredyty']
df['debt_to_income_ratio'] = debt / income

# Wykresy
plt.figure(figsize=(12, 6))
sns.distplot(df['ratio_expenses_income'], kde=False, bins=30)
plt.title('Rozkład wydatków do dochodów')
plt.xlabel('Stosunek wydatków do dochodów')
plt.ylabel('Liczba pracowników')
plt.show()
```

##### Wskaźnik równowagi budżetowej

```python
df['balance_ratio'] = (income - expenses - debt) / income

#Dodaj wykres slupkowy
plt.figure(figsize=(12, 6))
df['balance_ratio'].plot(kind='hist', bins=10)
plt.title('Rozkład wskaźnika bilansu')
plt.show()
```

##### Wskaźnik udziału w dochodzie dla poszczególnych kategorii wydatków

```python
def oblicz_udzial_w_dochodzie(df):
    wydatki_kolumny = ['wydatki_mieszkanie', 'wydatki_zywnosc', 'wydatki_transport', 'wydatki_edukacja']
    for kolumna in wydatki_kolumny:
        nowa_kolumna = f'{kolumna}_udzial'
        df[nowa_kolumna] = (df[kolumna] / df['roczny_dochod'])
    
    return df

df = oblicz_udzial_w_dochodzie(df)
df.head()
```

##### Korelacja miedzy rocznym dochodem a zadluzeniem

```python
df['inwestycje'] = df['inwestycje_akcje'] + df['inwestycje_nieruchomosci']
correlation = df['roczny_dochod'].corr(df['inwestycje'])
print("Korelacja między rocznym dochodem a inwestycjami:", correlation)

plt.figure(figsize=(12, 6))
plt.scatter(df['roczny_dochod'], df['inwestycje'])
plt.title('Korelacja między rocznym dochodem a inwestycjami')
z = np.polyfit(df['roczny_dochod'], df['inwestycje'], 1)
p = np.poly1d(z)
plt.plot(df['roczny_dochod'],p(df['roczny_dochod']),"r--")
plt.xlabel('Roczny dochód')
plt.ylabel('Inwestycje')
plt.show()
```

##### Korelacja miedzy zadluzeniem a rocznym dochodem

```python
df['debt'] = df['kredyty'] + df['karty_kredytowe']
correlation = df['roczny_dochod'].corr(df['debt'])
print("Korelacja między rocznym dochodem a zadluzeniem:", correlation)

plt.figure(figsize=(12, 6))
plt.scatter(df['roczny_dochod'], df['debt'])
plt.title('Korelacja między rocznym dochodem a zadluzeniem')
z = np.polyfit(df['roczny_dochod'], df['debt'], 1)
p = np.poly1d(z)
plt.plot(df['roczny_dochod'],p(df['roczny_dochod']),"r--")
plt.xlabel('Roczny dochód')
plt.ylabel('Zadluzenie')
plt.show()
```
