```python
#Download data from Yahoo Finance

import yfinance as yf

# Pobranie danych dla indeksu DAX
dax = yf.Ticker("^GDAXI")

# Pobranie danych historycznych, np. od 1 stycznia 2020 roku
dane_dax = dax.history(start="2020-01-01", end="2024-10-12")

dane_dax.reset_index(inplace=True)
dane_dax = dane_dax[['Date', 'Open']]

# Wyświetlenie danych
print(dane_dax)
```

```python
# Pobranie danych dla Bitcoina (BTC-USD)
bitcoin = yf.Ticker("BTC-USD")

# Pobranie danych historycznych, np. od 1 stycznia 2020 roku
dane_bitcoin = bitcoin.history(start="2020-01-01", end="2024-10-12")

# Resetowanie indeksu, aby kolumna 'Date' była dostępna
dane_bitcoin.reset_index(inplace=True)
dane_bitcoin = dane_bitcoin[['Date', 'Open']]

# Wyświetlenie danych
print(dane_bitcoin)
```

```python
# Pobranie danych dla Złota (Gold Futures - GC=F)
zloto = yf.Ticker("GC=F")

# Pobranie danych historycznych, np. od 1 stycznia 2020 roku
dane_zloto = zloto.history(start="2020-01-01", end="2024-10-12")

dane_zloto.reset_index(inplace=True)
dane_zloto = dane_zloto[['Date', 'Open']]

# Wyświetlenie danych
print(dane_zloto)
```

```python
#Funkcja zostawiajaca ostatnią wartość po przecinku z kolumny 0 dataframe
def last_value(data):
    desired = data.iloc[:, 0]
    desired = desired.str.split(',').str[-1]
    data.iloc[:, 0] = desired
    data.rename(columns={data.columns[0]: 'Country'}, inplace=True)

    return data

last_value(pkb)

pkb = pkb[pkb['Country'] == 'DE']

pkb
```
